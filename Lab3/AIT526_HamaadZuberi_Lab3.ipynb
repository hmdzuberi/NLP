## Task 1: Named Entity Recognition (NER)
### 1.1 Copy the code examples to scrape the webpage in BeautifulSoup

```python
import bs4 as bs
import urllib.request
import re

def _scrape_webpage(url):
    """
    Use BeautifulSoup to scrape a webpage and return the text.
    """
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}
    req = urllib.request.Request(url, headers=headers)
    scraped_text = urllib.request.urlopen(req).read()
    parsed_text = bs.BeautifulSoup(scraped_text, 'html.parser')
    paragraphs = parsed_text.find_all('p')
    formatted_text = ""
    
    for para in paragraphs:
        formatted_text += para.text
        
    return formatted_text

raw_text = _scrape_webpage('https://apnews.com/article/india-eu-modi-ursula-von-der-leyen-309ab4795ad0206b66fe20bef5ca9a92')

# Remove references (e.g., [1], [2], etc.) from the text
raw_text = re.sub(r'\[[0-9]*\]', ' ', raw_text)

# Remove extra spaces from the text
raw_text = re.sub(r'\s+', ' ', raw_text)

print(raw_text)
```

### 1.2 Write the code for NER in SpaCy


```python
import spacy
nlp = spacy.load('en_core_web_sm')

from spacy import displacy
from collections import Counter
from pprint import pprint

text = nlp(raw_text)

# This code prints each token in the text along with its IOB (Inside-Outside-Beginning) tag and entity type
pprint([(X, X.ent_iob_, X.ent_type_) for X in text])
```

#### 1.2.1 Count all the named entities in the document

```python
# Print each named entity in the text along with its label
pprint([(X.text, X.label_) for X in text.ents])

# Extract the labels of all named entities
labels = [x.label_ for x in text.ents]

from collections import Counter

# Count the frequency of each label
Counter(labels)
```

#### 1.2.2 Count the most frequent tokens for the entire document

```python
# Get most frequent tokens
items = [x.text for x in text.ents]
Counter(items).most_common(5)
```

#### 1.2.3 Pick a random integer K using Python random module, then pick three consecutive sentences starting with Kth, and print these sentences.
Note that you must make sure all picked sentences are in the document.

```python
import random

sentences = list(text.sents)
K = random.randint(0, len(sentences) - 3)

# Select three consecutive sentences starting with the Kth sentence
selected_sentences = sentences[K:K+3]

# Print the selected sentences
for sentence in selected_sentences:
    print(sentence)
```

#### 1.2.4 Extract part-of-speech and lemmatize these consecutive sentences

```python
for sentence in selected_sentences:
    for token in sentence:
        print(f"Token: {token.text}, POS: {token.pos_}, Lemma: {token.lemma_}")
```

#### 1.2.5 Get and print the entity annotation for each token of the Kth sentence

```python
# Get the Kth sentence
kth_sentence = sentences[K]

# Print the entity annotation for each token in the Kth sentence
for token in kth_sentence:
    print(f"Token: {token.text}, Entity: {token.ent_type_ if token.ent_type_ else 'None'}")
```

#### 1.2.6 Visualize the entities and dependencies of Kth sentence

```python
# Visualize the entities
displacy.render(kth_sentence, style='ent', jupyter=True)

# Visualize the dependencies
displacy.render(kth_sentence, style='dep', jupyter=True)
```

## Task 2: De-identification
### 2.1 De-identify all person names (PERSON) in the webpage document with [REDACTED]

```python
# Create a copy of the raw_text to perform de-identification
deidentified_text = raw_text

# Replace all PERSON entities with [REDACTED]
for ent in text.ents:
    if ent.label_ == 'PERSON':
        deidentified_text = deidentified_text.replace(ent.text, '[REDACTED]')

print(deidentified_text)
```
